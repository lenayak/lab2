```python
import matplotlib.pyplot as plt
import re
```

```python
threads = [2, 5, 10, 16, 20]
for thread in threads:
  file = open(f'average_times_{thread}_threads.txt', 'r')
  file = file.read()
  sizes_and_times = re.findall(r"[-+]?\d*\.\d+|\d+", file)

  sizes = sizes_and_times[0:len(sizes_and_times):2]
  times = sizes_and_times[1:len(sizes_and_times):2]

  sizes = [float(i) for i in sizes]
  times = [float(i) for i in times]

  plt.plot(sizes, times, marker='o', linestyle='-', label=f'{thread}')
  plt.title('Среднее время')
  plt.xlabel('Размеры матриц')
  plt.ylabel('Время')
  plt.legend()
```
